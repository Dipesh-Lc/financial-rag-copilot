```python
import sys
sys.path.insert(0, '..')
from app.embeddings.embedding_benchmark import benchmark_model, CANDIDATE_MODELS
```

```python
# Small benchmark corpus — replace with real chunks after ingestion
corpus = [
    'Apple faces intense competition in smartphone markets from Android manufacturers.',
    'Apple supply chain is concentrated in Asia with risks of disruption.',
    'Microsoft Azure cloud revenue grew 20% driven by AI services.',
    'Microsoft faces regulatory scrutiny in EU for bundling practices.',
    'Alphabet advertising revenue declined due to macro headwinds.',
    'Google Cloud grew 28% year over year reaching 9.4 billion dollars.',
    'Apple regulatory risks include EU Digital Markets Act compliance.',
    'Microsoft cybersecurity incidents disclosed in annual filing.',
]

queries = [
    'What competition risks does Apple face?',
    'How did Microsoft cloud revenue perform?',
    'What regulatory risks does Google face?',
]

# relevant[i] = list of corpus indices relevant to queries[i]
relevant_indices = [[0, 1, 6], [2, 7], [4, 5]]
```

```python
import pandas as pd

rows = []
for model_name in CANDIDATE_MODELS:
    print(f'Benchmarking {model_name}...')
    result = benchmark_model(model_name, queries, corpus, relevant_indices)
    rows.append(result)

pd.DataFrame(rows).sort_values('hit_rate_at_5', ascending=False)
```
